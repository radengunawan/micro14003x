```python
import matplotlib.pyplot as plt

# Data points for the indifference curve
pizza = [3, 4, 6]  # Quantity of Pizza (X-axis)
soda = [6, 4, 2]   # Quantity of Soda (Y-axis)

# Plotting the points
plt.plot(pizza, soda, 'bo', label='Indifference Points')

# Drawing the indifference curve
plt.plot(pizza, soda, 'b-', label='Indifference Curve')

# Labeling the axes
plt.xlabel('Slices of Pizza')
plt.ylabel('Cans of Soda')

# Adding a title
plt.title('Indifference Curve for Pizza and Soda')

# Adding a legend
plt.legend()

# Displaying the graph
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Utility levels
utility_levels = [2, 4, 6]

# Plotting the indifference curves
for u in utility_levels:
    # Points for the L-shaped curve
    CF = [u, u]
    BF = [0, u]

    plt.plot(CF, BF, 'b-')  # Vertical part of the L
    plt.plot([0, u], [u, u], 'b-')  # Horizontal part of the L
    plt.plot(u, u, 'bo')  # Mark the corner of the L

# Labeling the axes
plt.xlabel('Coca-Cola (C_F)')
plt.ylabel('Biscuits (B_F)')

# Adding a title
plt.title('Indifference Curves for U_F = min(C_F, B_F)')

# Adding grid for clarity
plt.grid(True)

# Displaying the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the utility function and utility level
def utility(x, y):
    return 2 * x + 3 * y

# Define the utility level for the indifference curve
utility_level = 12

# Create values for x and compute corresponding y values for the indifference curve
x_values = np.linspace(0, 6, 100)
y_values = (utility_level - 2 * x_values) / 3

# Plotting the indifference curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'U(x, y) = {utility_level}')
plt.title('Indifference Curve for U(x, y) = 2x + 3y')
plt.xlabel('Quantity of Coffee (x)')
plt.ylabel('Quantity of Tea (y)')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the utility function for U(x, y) = min(x, y)
def utility_min(x, y):
    return np.minimum(x, y)

# Define the utility level for the indifference curve
utility_level_min = 3

# Create values for x and compute corresponding y values for the indifference curve
x_values = np.linspace(0, 6, 100)
y_values = np.where(x_values < utility_level_min, x_values, utility_level_min)

# Plotting the indifference curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'U(x, y) = {utility_level_min}')
plt.plot(y_values, x_values, label=f'U(x, y) = {utility_level_min}', linestyle='--')
plt.title('Indifference Curve for U(x, y) = min(x, y)')
plt.xlabel('Quantity of Coffee (x)')
plt.ylabel('Quantity of Tea (y)')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define utility levels
utility_levels = [3, 4, 5]

# Plotting the indifference curves for different utility levels
plt.figure(figsize=(8, 6))

for level in utility_levels:
    x_values = np.linspace(0, 6, 100)
    y_values = np.where(x_values < level, x_values, level)

    plt.plot(x_values, y_values, label=f'U(x, y) = {level}')
    plt.plot(y_values, x_values, linestyle='--')

plt.title('Indifference Curves for U(x, y) = min(x, y)')
plt.xlabel('Quantity of Coffee (x)')
plt.ylabel('Quantity of Tea (y)')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
